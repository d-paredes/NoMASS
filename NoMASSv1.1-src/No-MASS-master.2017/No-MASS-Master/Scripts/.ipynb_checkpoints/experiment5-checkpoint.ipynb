```python
weekly_cost = ad_AL.groupby(by='nsim').Building0_Appliance1_cost.sum()
weekly_poweruse = ad_AL.groupby(by='nsim').Building0_Appliance1_recieved.sum()
```

```python
weekly_df_AL = pd.DataFrame({'cost': weekly_cost, 'poweruse': weekly_poweruse})
```

```python
fig1=plt.figure(figsize = (14,4))
sns.boxplot(weekly_df_AL, orient = 'h')
plt.xlabel('Weekly data (for 50 simulations)')
```

### Self conusmption

```python
dif = ad_AL.Building0_Appliance10_supplied  -  ad_AL.Building0_Appliance1_recieved
ad['gridImport'] = dif.where(dif<0,0)
ad['pvExport'] = dif.where(dif>0,0)
ad['pvUse'] = ad.Building0_Appliance10_supplied - ad['pvExport'] 
```

```python
# totals
grouped = ad.groupby('nsim')
weekly_demand = grouped.Building0_Appliance1_recieved.sum()
weekly_pv     = grouped.Building0_Appliance10_supplied.sum()
weekly_grid   = grouped.gridImport.sum()
weekly_pvexp  = grouped.pvExport.sum()
weekly_pvuse  = grouped.pvUse.sum()
```

```python
weekly_df_AL['self_cons_index'] = weekly_pvuse/weekly_demand *100
weekly_df_AL['self_cons_potential'] = weekly_pv/weekly_demand *100
```

```python
fig2 = plt.figure(figsize = (14,3))
sns.boxplot(weekly_df_AL.self_cons_index, orient = 'h')
plt.xlabel('self consumption index (%)')

# sns.boxplot(self_consumption_index)
```

### Matrix shifts

```python
matrix_shifts =action.groupby(by=['Building0_Appliance1-_previous_state1','Building0_Appliance1-_action1']).size()


matrix_shifts =pd.DataFrame(matrix_shifts.reset_index())
matrix_shifts.columns = ['from_time', 'to_time', 'ntimes']

# sns.set()
figure31 = plt.figure(figsize=(10,10))
matrix_s = matrix_shifts.pivot("from_time", "to_time", "ntimes").fillna(0)


vmin, vmax = matrix_s.values.min(), matrix_s.values.max()
cmap = sns.cubehelix_palette(vmax, start=0.9, rot=0., as_cmap=True, light = 0.97)
sns.heatmap(matrix_s,  vmin=vmin, vmax=vmax,
            annot = True, linewidths=.5, cbar_kws={"shrink": .7},
            cmap=cmap,square=True,
            xticklabels = range(1,25), yticklabels=range(1,25))
```

```python
# matrix_shifts =action.groupby(by=['Building0_Appliance1-_state1','Building0_Appliance1-_action1']).count()
# matrix_shifts
# # matrix_shifts[matrix_shifts.from_time ==13]
```

```python
# Modify shifting times heatmap to periods of day, instead of hours.
def maptime(value):
    if 0 <= value < 6:
        return 'nightime'
    elif 6 <= value < 9:
        return 'morning'
    elif 9 <= value < 17:
        return 'suntime'
    else:
        return 'evening'
matrix_shifts['from_period'] = matrix_shifts.from_time.map(maptime)
matrix_shifts['to_period'] = matrix_shifts.to_time.map(maptime)
cmap = sns.cubehelix_palette(15, start=0.9, rot=0., as_cmap=True)

plt.figure(figsize=(6,6))
matrix_shifts_per = pd.DataFrame(matrix_shifts.groupby(['from_period','to_period']).size()).reset_index().rename(columns = {0:'ntimes'})
matrix_s_per = matrix_shifts_per.pivot('from_period','to_period','ntimes').fillna(0)
to_suntime_shifts = matrix_s_per.suntime.sum()/matrix_s_per.sum().sum() * 100

sorted_cols = ['nightime','morning','suntime','evening']
sns.heatmap(matrix_s_per[sorted_cols].sort_index(by=sorted_cols), 
            annot = True, linewidths=.5, cbar_kws={"shrink": .8},
            cmap=cmap,square=True)
plt.title('%.1f per cent of shifts to suntime'%to_suntime_shifts)
figure32 = plt.gcf()
```
